```python
import pandas as pd

# 파일 경로 지정
basic_path = "crawling_results/rndjob_basic_20250425_161103.csv"
detail_path = "crawling_results/rndjob_detail_20250425_161103.csv"

# CSV 파일 불러오기
basic_df = pd.read_csv(basic_path)
detail_df = pd.read_csv(detail_path)
```

```python
# 컬럼명 확인
print("📌 Basic 컬럼들:", basic_df.columns.tolist())
print("📌 Detail 컬럼들:", detail_df.columns.tolist())

# 각 데이터프레임의 상위 5개 행 미리보기
display(basic_df.head())
display(detail_df.head())

# 데이터 크기 확인
print("기본정보 데이터 개수:", basic_df.shape)
print("상세정보 데이터 개수:", detail_df.shape)
```

```python
# 병합 가능한 키 컬럼 확인
common_cols = set(basic_df.columns).intersection(detail_df.columns)
print("공통 컬럼:", common_cols)

# 상세 URL을 기준으로 병합 가능성 판단
merged_df = pd.merge(basic_df, detail_df, on="상세정보_URL", how="inner")
print("병합된 행 개수:", merged_df.shape[0])
```

```python
# 각 컬럼 데이터 타입 및 결측값 비율 확인
print("📑 basic_df info:")
print(basic_df.info())
print(basic_df.isnull().sum())

print("\n📑 detail_df info:")
print(detail_df.info())
print(detail_df.isnull().sum())
```

```python
# 예시로 몇 개 컬럼만 출력해봄
columns_to_check = ['지원자격', '근무조건', '등록일/마감일', '담당업무', '자격사항', '근무환경']

for col in columns_to_check:
    if col in basic_df.columns:
        print(f"\n📝 {col} (기본 정보):")
        print(basic_df[col].dropna().sample(3, random_state=1).tolist())
    elif col in detail_df.columns:
        print(f"\n📝 {col} (상세 정보):")
        print(detail_df[col].dropna().sample(3, random_state=1).tolist())
```

```python
def print_unique_values(df, column_name):
    """
    특정 컬럼에서 중복을 제거한 고유값을 모두 출력하는 함수
    """
    if column_name not in df.columns:
        print(f"[오류] '{column_name}' 컬럼은 데이터프레임에 존재하지 않습니다.")
        return

    unique_vals = df[column_name].dropna().unique()
    print(f"\n📦 '{column_name}' 컬럼의 고유값 목록 ({len(unique_vals)}개):")
    for val in unique_vals:
        print("-", val)
```

```python
print_unique_values(merged_df, "지원자격")
print_unique_values(merged_df, "경력")
```

```python
import ast

def extract_work_location(row):
    """
    '근무환경' 컬럼에서 '근무지역' 값을 추출하는 함수
    """
    try:
        # 문자열을 리스트로 변환
        env_list = ast.literal_eval(row)
        # '근무지역 :' 로 시작하는 문자열 찾기
        for item in env_list:
            if item.startswith('근무지역'):
                return item.split(' : ')[1]  # '근무지역 : 서울(서초구)' → '서울(서초구)'
    except Exception as e:
        return None

# detail_df의 '근무환경' 컬럼에서 근무지역 추출
merged_df['근무지역'] = merged_df['근무환경'].apply(extract_work_location)

# 결과 확인
print(merged_df[['회사명', '근무지역']].dropna().head(10))
print(merged_df['근무지역'].value_counts())
```
